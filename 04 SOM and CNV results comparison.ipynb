```python
import pandas as pd
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import plotly.express as px
```

### Data

```python
### CNV data
hpv_pos_amp_top_drugBank_drug_candidates = pd.read_csv('Results/hpv_pos_amp_top_drugBank_drug_candidates.csv')
hpv_pos_del_top_drugBank_drug_candidates = pd.read_csv('Results/hpv_pos_del_top_drugBank_drug_candidates.csv')

hpv_neg_amp_top_drugBank_drug_candidates = pd.read_csv('Results/hpv_neg_amp_top_drugBank_drug_candidates.csv')
hpv_neg_del_top_drugBank_drug_candidates = pd.read_csv('Results/hpv_neg_del_top_drugBank_drug_candidates.csv')  
```

```python
### som data
hpv_pos_som_top_drugBank_drug_candidates = pd.read_csv('Results/hpv_pos_som_top_drugBank_drug_candidates.csv')
hpv_neg_som_top_drugBank_drug_candidates = pd.read_csv('Results/hpv_neg_som_top_drugBank_drug_candidates.csv')
```

### Compare results

#### HPV+

```python
### Venn diagram for hpv pos genes
print(f'Number of genes in HPV positive amplification genes: {hpv_pos_amp_top_drugBank_drug_candidates['GENE_TARGET'].nunique()}')
print(f'overlap between hpv positive amplification and deletion: {len(set(hpv_pos_amp_top_drugBank_drug_candidates["GENE_TARGET"]).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["GENE_TARGET"])))}')
print(f'Number of genes in HPV positive deletion genes: {hpv_pos_del_top_drugBank_drug_candidates['GENE_TARGET'].nunique()}')

print(f'Number of genes in hpv positive somatic genes: {hpv_pos_som_top_drugBank_drug_candidates['GENE_TARGET'].nunique()}')
print(f'overlap between hpv positive somatic and amplification: {len(set(hpv_pos_som_top_drugBank_drug_candidates["GENE_TARGET"]).intersection(set(hpv_pos_amp_top_drugBank_drug_candidates["GENE_TARGET"])))}')
print(f'overlap between hpv positive somatic and deletion: {len(set(hpv_pos_som_top_drugBank_drug_candidates["GENE_TARGET"]).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["GENE_TARGET"])))}')

print(f'Overlap between hpv positive amplification and deletion and somatic: {len(set(hpv_pos_amp_top_drugBank_drug_candidates["GENE_TARGET"]).intersection(set(hpv_pos_som_top_drugBank_drug_candidates["GENE_TARGET"])).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["GENE_TARGET"])))}')
```

```python
### Venn diagram for hpv pos drugs

print(f'Number of drugs in HPV positive amplification drugs: {hpv_pos_amp_top_drugBank_drug_candidates['DRUG'].nunique()}')
print(f'overlap between hpv positive amplification and deletion: {len(set(hpv_pos_amp_top_drugBank_drug_candidates["DRUG"]).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["DRUG"])))}')
### those in hpv pos del not in amp
print('Number of Drugs in hpv pos del not in amp:')
print(len(set(hpv_pos_del_top_drugBank_drug_candidates['DRUG']) - set(hpv_pos_amp_top_drugBank_drug_candidates['DRUG'])))
### those in hpv pos amp not in del
print('Number of Drugs in hpv pos amp not in del:')
print(len(set(hpv_pos_amp_top_drugBank_drug_candidates['DRUG']) - set(hpv_pos_del_top_drugBank_drug_candidates['DRUG'])))


print(f'Number of drugs in HPV positive deletion drugs: {hpv_pos_del_top_drugBank_drug_candidates['DRUG'].nunique()}')

print(f'Number of drugs in hpv positive somatic drugs: {hpv_pos_som_top_drugBank_drug_candidates['DRUG'].nunique()}')
### only in som not in amp or del
print('Number of Drugs in hpv pos som not in amp or del:')
print(len(set(hpv_pos_som_top_drugBank_drug_candidates['DRUG']) - set(hpv_pos_amp_top_drugBank_drug_candidates['DRUG']) - set(hpv_pos_del_top_drugBank_drug_candidates['DRUG'])))
print(f'overlap between hpv positive somatic and amplification: {len(set(hpv_pos_som_top_drugBank_drug_candidates["DRUG"]).intersection(set(hpv_pos_amp_top_drugBank_drug_candidates["DRUG"])))}')
print(f'overlap between hpv positive somatic and deletion: {len(set(hpv_pos_som_top_drugBank_drug_candidates["DRUG"]).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["DRUG"])))}')

print(f'Overlap between hpv positive amplification and deletion and somatic: {len(set(hpv_pos_amp_top_drugBank_drug_candidates["DRUG"]).intersection(set(hpv_pos_som_top_drugBank_drug_candidates["DRUG"])).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["DRUG"])))}')
```

```python
hpv_pos_del_top_drugBank_drug_candidates
```

```python
### show drugs of overlap
hpv_pos_drug_overlap = set(hpv_pos_amp_top_drugBank_drug_candidates["DRUG"]).intersection(
	set(hpv_pos_som_top_drugBank_drug_candidates["DRUG"])
).intersection(set(hpv_pos_del_top_drugBank_drug_candidates["DRUG"]))

top_hpv_pos_overlap_drug = hpv_pos_amp_top_drugBank_drug_candidates[
	hpv_pos_amp_top_drugBank_drug_candidates["DRUG"].isin(hpv_pos_drug_overlap)
]
top_hpv_pos_overlap_drug.drop_duplicates(subset=('DRUG', 'GENE_TARGET'), inplace=True)
top_hpv_pos_overlap_drug
```

```python
hpv_pos_drug_overlap
```

```python
### combine top candidate information from sources
hpv_top_drug_candidate_information = pd.concat(
    [hpv_pos_amp_top_drugBank_drug_candidates, hpv_pos_del_top_drugBank_drug_candidates, hpv_pos_som_top_drugBank_drug_candidates],
    ignore_index=True
)
hpv_top_drug_candidate_information = hpv_top_drug_candidate_information[hpv_top_drug_candidate_information['DRUG'].isin(hpv_pos_drug_overlap)]

### aggregate this to show how many genes each drug targets
agg_hpv_top_drug_candidate_information = hpv_top_drug_candidate_information.groupby('DRUG').agg({
    'GENE_TARGET': lambda x: ', '.join(set(x)),
    'CONNECTED_TO': lambda x: ','.join(set(x)),
    'drug_hypergeom_fdr': 'min',

    'drug_empirical_fdr': 'min',
}).reset_index()

agg_hpv_top_drug_candidate_information['num_targeted_genes'] = agg_hpv_top_drug_candidate_information['GENE_TARGET'].apply(lambda x: len(x.split(', ')))
agg_hpv_top_drug_candidate_information['num_connected_to'] = agg_hpv_top_drug_candidate_information['CONNECTED_TO'].apply(lambda x: len(set(x.split(','))))
agg_hpv_top_drug_candidate_information = agg_hpv_top_drug_candidate_information.sort_values(by=['drug_hypergeom_fdr', 'drug_empirical_fdr'], ascending=[True, True]).reset_index(drop=True)
agg_hpv_top_drug_candidate_information
```
